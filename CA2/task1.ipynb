2. Show the type of each data column. Some columns are of type categorical and some are of type numerical from the pandas library info. To process the non-numerical columns, one of the possible methods is labeling; in such a way that each of the categories is replaced by a number. For example, in this dataset, there is a categorical column named sex, which includes values Male and Female. Modify the values of this column, so that each of these models is mapped to one of the numbers in the range [0,1].

Here, in order to demonstrate the non-numerical values using numbers:

in 'sex' column, 'male' is set to 0 and 'female' to 1,
in 'embarked' column, 'S' is set to 0, 'Q' to 1, and 'C' to 2,
in 'class' column, 'First' is set to 1, 'Second' to 2, and 'Third' to 3,
in 'who' column, 'man' is set to 0, 'woman' to 1, and 'child' to 2,
in 'adult_male' column, 'True' is set to 1 and 'False' to 0, 
in 'deck' column, 'A' is set to 1, 'B' to 2, 'C' to 3, 'D' to 4, 'E' to 5, 'F' to 6, and 'G' to 7,
in 'embark_town' column, 'Southampton' is set to 0, 'Queentown' to 1, and 'Cherbourg' to 2,
in 'alive' column, 'yes' is set to 1 and 'no' to 0, and
in 'alone' column, 'True' is set to 1 and 'False' to 0. 

```python
df = data_set.copy()

df['sex'] = df['sex'].map({'male' : 0, 'female' : 1})
df['embarked'] = df['embarked'].map({'S' : 0, 'Q' : 1, 'C' : 2})
df['class'] = df['class'].map({'First' : 1, 'Second' : 2, 'Third' : 3})
df['who'] = df['who'].map({'man' : 0, 'woman' : 1, 'child' : 2})
df['adult_male'] = df['adult_male'].map({False : 0, True : 1})
df['deck'] = df['deck'].map({'A' : 1,'B' : 2,'C' : 3,'D' : 4,'E' : 5,'F' : 6,'G' : 7})
df['embark_town'] = df['embark_town'].map({'Southampton' : 0, 'Queentown' : 1, 'Cherbourg' : 2})
df['alive'] = df['alive'].map({'no' : 0, 'yes' : 1})
df['alone'] = df['alone'].map({False : 0, True : 1})

df.info()
print('~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~')
print(df.head())
print('~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~')
print(df.tail())
print('~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~')
df.describe()
```

 3. Generate a heatmap of the correlation matrix for numerical features in the dataset.
 - Correlation Matrix:
    - It's a tabular representation illustrating the correlation between pairs of variables within a dataset.
    - Correlation indicates the degree of association or dependence between two variables.
    - A value of 1 signifies a perfect positive correlation, implying that as one variable increases, the other increases proportionally.
    - Conversely, a value of -1 indicates a perfect negative correlation, where one variable increases as the other decreases.
    - A value of 0 suggests no correlation between the variables.

- Heatmap Visualization:
    - Heatmaps utilize color gradients to visualize the correlation values present in the matrix.
    - Typically, warmer colors like red denote positive correlations, with darker shades indicating stronger correlations.
    - Cooler tones such as blue signify negative correlations, with deeper shades indicating stronger associations.
    - Values closer to 0 are often represented in white or yellow, indicating weak or negligible correlations.

```python
correlation_matrix = df.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix Heatmap')
plt.show()
```

8. Identify columns with missing values. Propose and apply a strategy for handling these missing values.

```python
missing_columns = data_set.columns[data_set.isna().any()].tolist()

print("Columns with missing values:", missing_columns)
```

Handling missing values in a dataset is a crucial step in data preprocessing. Here's a strategy for handling missing values in both numerical and categorical features:

1. **Identify Missing Values**: Before handling missing values, it's important to identify them. In Python, missing values are often represented as `NaN`, `None`, or other placeholders.

2. **For Numerical Features**:
   - **Imputation**: Replace missing numerical values with a measure of central tendency such as mean, median, or mode.
   - **Advanced Imputation**: Replace it with values by making use of information from other columns. Use more advanced imputation techniques such as K-nearest neighbors (KNN) imputation.

3. **For Categorical Features**:
   - **Imputation**: Replace missing categorical values with the mode (most frequent value) of the respective feature.
    
   - **Create a New Category**: You can create a new category (e.g., "Unknown" or "Other") and assign missing categorical values to this category.


4. **Consider Dropping Missing Values**:
   - This is the simplest approach. We can drop rows with any missing values (df.dropna()) or drop entire columns with a high percentage of missing values. However, this can lead to data loss.


```python
# Replace the missing values in 'age' column for female passengers with the female median age
median_age = data_set[(data_set['sex'] == 'female')]['age'].median()
data_set.loc[data_set['sex'] == 'female', 'age'] = data_set.loc[data_set['sex'] == 'female', 'age'].fillna(median_age)

# Replace the missing values in 'age' column for male passengers with the male median age
median_age = data_set[(data_set['sex'] == 'male')]['age'].median()
data_set.loc[data_set['sex'] == 'male', 'age'] = data_set.loc[data_set['sex'] == 'male', 'age'].fillna(median_age)

# Replace the missing values in 'deck' column with the mode of deck class
most_frequent_deck = data_set['deck'].mode()[0]
data_set['deck'].fillna(most_frequent_deck, inplace=True)

# Replace the missing values in 'embarked' column with the mode od embarked class
most_frequent_embarked = data_set['embarked'].mode()[0]
data_set['embarked'].fillna(most_frequent_embarked, inplace=True)

# Replace the missing values in 'embark_town' column with the mode of embark_town class
most_frequent_embark_town = data_set['embark_town'].mode()[0]
data_set['embark_town'].fillna(most_frequent_embark_town, inplace=True)
```
