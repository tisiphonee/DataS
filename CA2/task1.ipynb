Question 2

Here, in order to demonstrate the non-numerical values using numbers:

in 'sex' column, 'male' is set to 0 and 'female' to 1,
in 'embarked' column, 'S' is set to 0, 'Q' to 1, and 'C' to 2,
in 'class' column, 'First' is set to 1, 'Second' to 2, and 'Third' to 3,
in 'who' column, 'man' is set to 0, 'woman' to 1, and 'child' to 2,
in 'adult_male' column, 'True' is set to 1 and 'False' to 0, 
in 'deck' column, 'A' is set to 1, 'B' to 2, 'C' to 3, 'D' to 4, 'E' to 5, 'F' to 6, and 'G' to 7,
in 'embark_town' column, 'Southampton' is set to 0, 'Queentown' to 1, and 'Cherbourg' to 2,
in 'alive' column, 'yes' is set to 1 and 'no' to 0, and
in 'alone' column, 'True' is set to 1 and 'False' to 0. 



```python

df['sex'] = df['sex'].map({'male' : 0, 'female' : 1})
df['embarked'] = df['embarked'].map({'S' : 0, 'Q' : 1, 'C' : 2})
df['class'] = df['class'].map({'First' : 1, 'Second' : 2, 'Third' : 3})
df['who'] = df['who'].map({'man' : 0, 'woman' : 1, 'child' : 2})
df['adult_male'] = df['adult_male'].map({False : 0, True : 1})
df['deck'] = df['deck'].map({'A' : 1,'B' : 2,'C' : 3,'D' : 4,'E' : 5,'F' : 6,'G' : 7})
df['embark_town'] = df['embark_town'].map({'Southampton' : 0, 'Queentown' : 1, 'Cherbourg' : 2})
df['alive'] = df['alive'].map({'no' : 0, 'yes' : 1})
df['alone'] = df['alone'].map({False : 0, True : 1})

df.info()
print('~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~')
print(df.head())
print('~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~')
print(df.tail())
print('~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~')
df.describe()
```

```python
correlation_matrix = df.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix Heatmap')
plt.show()
```

As it stands, the values of the main diagonal are all 1 because it's the correlation of two identical set of data. The correlation of the two columns, alive and survived, the two columns, embarked and embark_town, and the two columns, pclass and classm are also 1 because the dataset were identical but labelled different, so we can safely remove them.

```python
df = df.drop('pclass', axis=1)
df = df.drop('embark_town', axis=1)
df = df.drop('alive', axis=1)

df.info()
print('~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~')
print(df.head())
print('~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~')
print(df.tail())
print('~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~')
df.describe()

correlation_matrix = df.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
survived_count = df[df['survived'] == 1].count()
print(f"Number of passengers who survived: {survived_count.iloc[0]}")

female_over_30_count = df[(df['sex'] == 1) & (df['age'] > 30)].count()
print(f"Number of female passengers aged more than 30: {female_over_30_count.iloc[0]}")

cherbourg_high_fare_passengers = df[(df['embarked'] == 2) & (df['fare'] > 100)].count()
print(f"Passengers who embarked from Cherbourg and paid fare > $100: {cherbourg_high_fare_passengers.iloc[0]}")
```

Now, let's handle the missing data:

There are about 180 values missing in the column "age". the best strategy is to use the mean value to replace the missing data.

```python
# unknown_value = -1
# df['age'] = df['age'].fillna(unknown_value)
```

There are only 2 values missing in the column "embarked". the best strategy is to use the value with the mode (most frequent value) of the available data.

```python
mode_embarked = df['embarked'].mode()[0]
df['embarked'] = df['embarked'].fillna(mode_embarked)
```

The majority of data in "deck" column is missing. The best strategy, in my opinion, is to totally get rid of this column, because it's not used in data analysis.


```python
df = df.drop('deck', axis=1)
```

```python
survival_rate_by_class = df.groupby('class')['survived'].mean()
plt.bar(survival_rate_by_class.index, survival_rate_by_class.values)
plt.xlabel('Class')
plt.ylabel('Survival Rate')
plt.title('Proportion of Passengers that Survived by Class')
plt.xticks([1, 2, 3])
plt.show()
```

```python
survived = df[df['survived'] == 1]
not_survived = df[df['survived'] == 0]
survived_age = survived['age'].values
not_survived_age = not_survived['age'].values
bins = np.linspace(min(min(survived_age), min(not_survived_age)), max(max(survived_age), max(not_survived_age)), num = 10)
plt.hist(survived['age'], bins=bins, alpha=0.5, label='Survived', color="green")
plt.hist(not_survived['age'], bins=bins, alpha=0.5, label='Not Survived', color="gray")
plt.xlabel('Age')
plt.ylabel('Count')
plt.title('Age Distribution of Passengers (Survived vs Not Survived)')
plt.legend()
plt.show()
```

```python


survived_count, _ = np.histogram(survived_age, bins=bins)
not_survived_count, _ = np.histogram(not_survived_age, bins=bins)
 
total_survived = len(survived_age)
total_not_survived = len(not_survived_age)

survived_percentages = survived_count / total_survived * 100
not_survived_percentages = not_survived_count / total_not_survived * 100

plt.bar(bins[:-1], survived_percentages, width=np.diff(bins), alpha=0.5, color="green", label='Survived')
plt.bar(bins[:-1], not_survived_percentages, width=np.diff(bins), alpha=0.5,  color="gray", label='Not Survived')
plt.xlabel('Age')
plt.ylabel('Percentage')
plt.legend()
plt.show()
```

```python
plt.scatter(survived['age'], survived['fare'], color='green', alpha=0.5, label='Survived')
plt.scatter(not_survived['age'], not_survived['fare'], color='gray', alpha=0.5, label='Not Survived')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Relationship between Age and Fare Paid (Color-coded by Survival)')
plt.legend()
plt.show()
```

```python
grouped_data = df.groupby(['class', 'survived'])['fare'].mean().unstack()
class_labels = grouped_data.index
x = np.arange(len(class_labels))
bar_width = 0.35
survived_fares = grouped_data[1]
not_survived_fares = grouped_data[0]
fig, ax = plt.subplots()
rects1 = ax.bar(x - bar_width/2, survived_fares, bar_width, label='Survived')
rects2 = ax.bar(x + bar_width/2, not_survived_fares, bar_width, label='Not Survived')
ax.set_xlabel('Class')
ax.set_ylabel('Average Fare')
ax.set_title('Average Fare Paid by Passengers (Grouped by Class and Survival Status)')
ax.set_xticks(x)
ax.set_xticklabels(class_labels)
ax.legend()
plt.tight_layout()
plt.show()
```
