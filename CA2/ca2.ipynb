 3. Generate a heatmap of the correlation matrix for numerical features in the dataset.

```python
# Select only the numerical features from the DataFrame
numerical_features = data_set.select_dtypes(include=['float64', 'int64'])

# Compute the correlation matrix
correlation_matrix = numerical_features.corr()

# Plot the correlation matrix as a heatmap using matplotlib
plt.figure(figsize=(10, 8))
plt.imshow(correlation_matrix, cmap='coolwarm', interpolation='nearest')
plt.colorbar(label='Correlation')
plt.title('Correlation Matrix of Numerical Features')
plt.xticks(range(len(correlation_matrix.columns)), correlation_matrix.columns, rotation=45)
plt.yticks(range(len(correlation_matrix.index)), correlation_matrix.index)
plt.tight_layout()
plt.show()
```

```python
survived_passengers_data = data_set[data_set['survived'] == 1]
class_proportions = survived_passengers_data['class'].value_counts(normalize=True)


plt.figure(figsize=(8, 6))
class_proportions.plot(kind='bar', color='skyblue')
plt.title('Proportion of Saurvived Passengers by Class')
plt.xlabel('Class')
plt.ylabel('Proportion')
plt.xticks(rotation=0)  # Rotate x-axis labels if necessary
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
survived_passengers_age = np.array(survived_passengers_data['age'])
unsurvived_passengers_age = np.array(data_set[data_set['survived'] == 0]['age'])

fig, ax = plt.subplots(1, 2, figsize=(10, 6))

# Plot histogram for unsurvived passengers
ax[0].hist(unsurvived_passengers_age, bins=30, alpha=0.5, color='red', label='Unsurvived Passengers')

# Plot histogram for survived passengers
ax[1].hist(survived_passengers_age, bins=30, alpha=0.5, color='blue', label='Survived Passengers')

ax[0].set_title('Survived Passenger Ages')
ax[1].set_title('Unsurvived Passenger Ages')
#plt.title('Frequency Distribution of Survived Passenger Ages')
ax[0].set_xlabel('Age')
ax[1].set_xlabel('Age')
#plt.xlabel('Age')
ax[0].set_ylabel('Frequency')
ax[1].set_ylabel('Frequency')


plt.show()
```
